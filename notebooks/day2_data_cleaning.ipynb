```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../data/flight.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[["SUM_YR_1", "SUM_YR_2"]] = df[["SUM_YR_1", "SUM_YR_2"]].fillna(0)
```

```python
df["AGE"] = df["AGE"].fillna(df["AGE"].median())
```

```python
df["GENDER"] = df["GENDER"].fillna("Unknown")
```

```python
df["WORK_CITY"] = df["WORK_CITY"].fillna("Unknown")
df["WORK_PROVINCE"] = df["WORK_PROVINCE"].fillna("Unknown")
df["WORK_COUNTRY"] = df["WORK_COUNTRY"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df=df[(df["FLIGHT_COUNT"]>0)&((df["SUM_YR_1"] + df["SUM_YR_2"]) > 0)]
```

```python
df["FFP_DATE"] = pd.to_datetime(df["FFP_DATE"])
df["LOAD_TIME"] = pd.to_datetime(df["LOAD_TIME"])
df["FIRST_FLIGHT_DATE"] = pd.to_datetime(df["FIRST_FLIGHT_DATE"])
df["LAST_FLIGHT_DATE"] = pd.to_datetime(df["LAST_FLIGHT_DATE"])
```

```python
df["FFP_DATE"] = pd.to_datetime(df["FFP_DATE"], errors="coerce")
df["LOAD_TIME"] = pd.to_datetime(df["LOAD_TIME"], errors="coerce")
df["FIRST_FLIGHT_DATE"] = pd.to_datetime(df["FIRST_FLIGHT_DATE"], errors="coerce")
df["LAST_FLIGHT_DATE"] = pd.to_datetime(df["LAST_FLIGHT_DATE"], errors="coerce")
```

```python
df[["FFP_DATE", "LOAD_TIME", "FIRST_FLIGHT_DATE", "LAST_FLIGHT_DATE"]].isnull().sum()
```

```python
df["LOYALTY_DAYS"]=(df["LOAD_TIME"]-df["FFP_DATE"]).dt.days
```

```python
df["MONETARY"] = df["SUM_YR_1"] + df["SUM_YR_2"]
```

```python
df = df[df["LOYALTY_DAYS"].notna()]
```

```python
df[["LOYALTY_DAYS", "LAST_TO_END", "FLIGHT_COUNT", "MONETARY"]].describe()
```

```python
df.to_csv("../data/flight_cleaned.csv", index=False)
```
