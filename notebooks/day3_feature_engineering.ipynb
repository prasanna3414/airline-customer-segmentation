```python
import pandas as pd
import numpy as np 
```

```python
df=pd.read_csv("../data/flight_cleaned.csv")
df.head()
```

```python
features = df[
    ["LOYALTY_DAYS", "LAST_TO_END", "FLIGHT_COUNT", "MONETARY", "avg_discount"]
]
```

```python
features.columns = ["Loyalty", "Recency", "Frequency", "Monetary", "Discount"]
```

```python
features.describe()
```

```python
features = features.clip(
    lower=features.quantile(0.01),
    upper=features.quantile(0.99),
    axis=1
)
```

```python
features.describe()
```

```python
features.isnull().sum()
```

```python
features.to_csv("../data/flight_features.csv",index=False)
```
